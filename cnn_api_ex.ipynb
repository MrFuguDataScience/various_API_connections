```python
def what(start,finish):
    c=[]    
    for i in range(start,finish):
        sleep(randint(2,8))

#         url = 'https://search.api.cnn.io/content?q=bitcoin&size=1300&from=0&'+f"page={i}"+'&sort=newest'
        url = 'https://search.api.cnn.io/content?q=bitcoin&size=1300&'+ f"from={i}" +'&sort=newest'
#         print(url)
        json_data = requests.get(url).json()
        c.extend(json_data['result'])
        
#         q= pd.DataFrame(json_data['result']).append(json_data,ignore_index=True)
        
    return json_data['result'],c
# q
```
