```python
# /orders/{market}, where {market}= whatever your are looking for

base_url = "https://tradeogre.com/api/v1/orders/BTC-XMR"

# and 'page=1'
ex_2=requests.get(base_url).json()


from collections import defaultdict

bb=defaultdict(list)
for i in ex_2.items():
    if i[0] !='success': # ignore useless success message with get request
        # create default dictionary with list comprehension, 
        bb[i[0]].append([(k, v) for k, v in i[1].items()])

#         bb_[i[0]].append(i[1])
st_=pd.DataFrame(bb)

st_.head()


# pd.DataFrame(bb_)['buy']
```

```python
def parse_cols(column_names_list,df):
    
    for i in column_names_list:
        
        v=df.explode(i) #flatten lists
#create new columns to dataframe, converting df to list and using index from that df
        v[['price_perDollar_'+i, 'Qty_'+i]] = pd.DataFrame(v[i].tolist(),index=v.index)

    return v[['price_perDollar_'+i, 'Qty_'+i]]


sell=parse_cols(['sell'],pd.DataFrame(st_['sell']))

buy=parse_cols(['buy'],pd.DataFrame(st_['buy']))
# sell

buy
```
