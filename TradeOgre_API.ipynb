```python
def parse_cols(column_names_list,df):
    
    for i in column_names_list:
        
        v=df.explode(i)

        v[['price_perDollar_'+i, 'Qty_'+i]] = pd.DataFrame(v[i].tolist(),index=v.index)

    return v[['price_perDollar_'+i, 'Qty_'+i]]


sell=parse_cols(['sell'],pd.DataFrame(st_['sell']))

buy=parse_cols(['buy'],pd.DataFrame(st_['buy']))
sell
buy
```
